```python
x=np.array([5,4,3,2,1])
y=pow(x,3)
plt.axis([1,6,1,150])
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,label='y = x ** 3')
plt.legend()
```

```python
plt.plot(x,y,label='y = x ** 3')
plt.legend(loc=4,shadow=True)
```

```python
z=pow(x,4)
w=pow(x,5)
plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,w)
plt.title("The result")
plt.text(4.5,200,'x**3')
plt.text(4,400,'x**4')
plt.text(3.5,900,'x**5')
plt.show()
```
